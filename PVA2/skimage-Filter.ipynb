```python
import skimage
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
skimage.__version__
```

```python
from skimage import data
camera = data.camera()
im = skimage.io.imread('data/mushroom.png')
im = camera
%matplotlib inline
plt.imshow(im,cmap='gray');
im.shape
```

```python
if len(im.shape) >2 and im.shape[2]>1:
    im = np.mean(im,axis=2)
```

```python
def show(im):
    plt.imshow(im,cmap='gray');
    plt.axis('off')
```

```python
from skimage.filters import sobel
sobel_im = sobel(im)
show(sobel_im);
plt.savefig('sobel.png')
```

```python
from skimage.filters import prewitt
prewitt_im = prewitt(im)
show(prewitt_im)
plt.savefig('prewitt.png')
```

```python
show(np.abs(sobel_im - prewitt_im))
```

```python
from skimage import filters
roberts_im = filters.roberts(camera)
show(roberts_im)
```

```python
import skimage.feature
canny_im = skimage.feature.canny(im)
show(canny_im)
```

```python
canny_im = skimage.feature.canny(im,sigma=5,low_threshold=10,high_threshold=50)
show(canny_im)
```
