```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
def generateData(separation = 0.5, c1Samples = 100, c2Samples = 100):
    """
    Generate a 2d separable training data set.
    separation : minimal distance between any two point with different class labels
    """
    NSamples = c1Samples+c2Samples
    x1,y1,one = np.random.rand(c1Samples ,1),np.random.rand(c1Samples,1),np.ones((c1Samples,1))
    x2,y2,minusone= 1+separation+np.random.rand(c2Samples,1),np.random.rand(c2Samples,1),-np.ones((c2Samples,1))
    
    
    data = np.vstack((np.hstack((x1,y1,one)),np.hstack((x2,y2,minusone))))#,columns=['x1','x2','y'])
    X = data[:,0:2]-1.0
    y = data[:,2]
    return X,y,NSamples
```

```python
def plotSeparationLine(beta0,beta,xline,color = 'c',iIter=None):
    print('y=%2.2f%+2.2f x'%(-beta0/beta[1],-beta[0]/beta[1]))
    yline = (-beta0-xline*beta[0])/beta[1]
    #only plot if the line crosses the data boundary
    LA = np.logical_and(x2min < yline,yline <x2max )
    plt.ion()
    if np.sum(LA)>1:
        line = plt.plot(xline[LA],yline[LA],c=color)
        NrMisclassifications = np.sum(np.sign(beta0+np.dot(beta,X.T)) - y != 0)
        if NrMisclassifications == 0:
            print('FINISHED!')
            return NrMisclassifications,yline
        print('# misclassifications: %i'%NrMisclassifications)
        if iIter:
            plt.title('Iteration {0}'.format(iIter))
        plt.draw()
        x = 5

    plt.draw()
#plt.show(block=True)
    NrMisclassifications = np.sum(np.sign(beta0+np.dot(beta,X.T)) - y != 0)
    return NrMisclassifications,yline
```

```python
def showData(X,y):
    c1LA = y>0
    c2LA = y<0
    plt.scatter(X[c1LA,0],X[c1LA,1],c='b',label='y=+1')
    plt.scatter(X[c2LA,0],X[c2LA,1],c='r',label='y=-1')
    #x1min is the minimum x-coordinate of all points 
    x1min,x1max = np.min(X[:,0]),np.max(X[:,0])
    #x2max is the maximum y-coordinate of all points 
    x2min,x2max = np.min(X[:,1]),np.max(X[:,1])
    plt.legend()
    plt.ion()
    plt.xlim(x1min,x1max)
    plt.ylim(x2min,x2max)
    plt.draw()
    return (x1min,x1max),(x2min,x2max)
```

```python
NrIterations = 1000000,

c1Samples = 100
c2Samples = 100
rho=0.01 #learning rate

X,y,NSamples = generateData(c1Samples=c1Samples,c2Samples=c2Samples)
(x1min,x1max),(x2min,x2max) = showData(X,y)

beta = np.random.random(2)
#beta = X.ix[1,:]
beta0 = random.random()

indexList = np.arange(NSamples)
xline = np.linspace(x1min,x1max,1000)

line = None
yline = (-beta0-xline*beta[0])/beta[1]

#clip the separating line to the data region and plot it
LA = np.logical_and(x2min < yline,yline <x2max )
line = plt.plot(xline[LA],yline[LA],c='c')

NrMisclassifications=1    
iIter=0
while NrMisclassifications>0:
    iIter+=1
    MisclassificationLA = np.sign(beta0+beta.dot(X.T)) -y != 0

    #choose a random misclassified point
    ind = random.choice(indexList[MisclassificationLA])
    xrow = X[ind,:]
    yrow = y[ind]

    # y*beta
    if yrow*(beta.dot(xrow)+beta0) < 0: #if misclassified
        print('-----')
        print('xi = %2.2f,%2.2f\ny=%+2.2f'%(xrow[0],xrow[1],yrow))

        #beta update
        beta += rho*yrow*xrow
        beta0 += rho*yrow

        #plot the separation line
        z = 1.0*NrMisclassifications/NSamples
        NrMisclassifications,yline = plotSeparationLine(beta0,beta,xline,color=[0,1-z,z],iIter=iIter)
        if NrMisclassifications == 0:
            break
    else:
        raise AssertionError('That''s strange...apparently, the selection of misclassified points failed?')

LA = np.logical_and(x2min < yline,yline <x2max )
line = plt.plot(xline[LA],yline[LA],c='m')
if NrMisclassifications>0:
    print('MaxIterations reached.')
else:
    print('perfect linear separability is given by y=%2.2f%+2.2f x'%(-beta0/beta[1],-beta[0]/beta[1]))
    plt.title('Perceptron')
    x=5	
plt.show(block=True)
print('end.')
```
